### Normalising and splitting the data

```python
#normalizing our data
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
normed_data = pd.DataFrame(minmax.fit_transform(df), columns = df.columns)
#normed_data.head()
target = normed_data['Appliances']
features_data = normed_data.drop(['Appliances'], axis = 1)
```

```python
#splitting data
from sklearn.model_selection import train_test_split
#splitter = train_test_split()
x_train, x_test, y_train, y_test = train_test_split(features_data, target, test_size = 0.3, random_state = 42)
```

### Training and Prediction

```python
#training data
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
#making a prediction
pred = LR.predict(x_test)
pred
```

### Getting the error scores

```python
# A function to produce the error scores
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
def score(test,predicted):
    """Args 
              test: the original Y_test values
              predicted: the predicted Y values
              
       Returns
              a print of MAE, MSE, ResSS and Rsquared error
    """
    mae = mean_absolute_error(test, predicted)
    mse =  np.sqrt(mean_squared_error(test, predicted))
    r2 = r2_score(test, predicted)
    rss = np.sum(np.square(test - predicted))
    return (f'mae error: {round(mae,3)}, mse error: {round(mse,3)}, r squared error: {round(r2,3)}, rss error: {round(rss,3)}')
```

```python
score(y_test,pred)
```

### Regularisation

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values() #gives the weight with the columns name of the features
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```python
#importing L1 & L2 rgularisation
from sklearn.linear_model import Lasso, Ridge
lasso = Lasso(alpha=0.001)
ridge = Ridge(alpha=0.4)
lass = lasso.fit(x_train,y_train)
rig = ridge.fit(x_train, y_train)
```

```python
linear_model_weights = get_weights_df(LR, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
